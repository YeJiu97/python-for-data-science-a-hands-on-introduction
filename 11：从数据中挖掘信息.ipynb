# 关联法则

关联法则，也被称为关联规则挖掘，是数据挖掘领域的一种关键研究方法。这种方法用于发现存在于项目集合或对象集合之间的频繁模式、关联、相关性或因果结构。它的基本形式是形如X→Y的蕴涵式，其中X和Y分别被称为关联规则的前驱（先导或左部，LHS）和后继（后继或右部，RHS）。

例如，假设我们有一个购物篮数据分析的场景，那么“购买了尿布”可能会被关联到“购买了啤酒”，这就是一个关联规则的例子。这种关联关系可以帮助商家进行商品推荐，提高销售额。

关联规则分析通常包括两个主要步骤：频繁项集的产生和规则的生成。频繁项集是指出现频率超过某一阈值的项目集合，而规则则是基于这些频繁项集生成的。

关联规则中的两个重要指标是支持度和置信度。支持度可以理解为同时包含A和B的事务在所有事务中的比例，或者更为具体地，表示项集在数据集中出现的频率。例如，假设有个项集包含“苹果”，如果在8次交易中出现了4次，那么这个项集的支持度就是50%。

而置信度则是反映当X出现时，Y也出现的概率。以一个实际例子来说，如果我们有一个关联规则：“购买了尿布”就“购买了啤酒”，那么当我们看到有人购买了尿布时，我们有多大的把握认为他们也购买了啤酒呢？这就需要用到置信度来判断了。

需要注意的是，通过设定支持度和置信度的阈值，我们可以筛选出数据集中的频繁项集和关联规则。例如，如果设定支持度阈值为0.5，置信度阈值为0.7，那么可以找到所有支持度大于等于0.5，置信度大于等于0.7的关联规则。

假设现在的情况如下所示：

首先计算支持度：
```
support(curd) = curd / total = 25 / 100 = 0.25
```

支持度是一件或者多件商品的交易次数占总交易次数的比例。

然后是计算关联规则，支持度是包括先导和后续的交易次数占总交易次数的比列，因此，curd -> sourcream 关联规则的支持度的计算方式如下所示：

```
support(curd -> sour cream) = (curd & sour cream) / total = 20 / 100 = 0.2
```

支持度的范围为 0~1，表示商品出现的百分比。

关联规则的置信度是购买先导和后续的交易次数与购买先导的交易次数的比列。

计算如下所示：

```
confidence(curd -> sour cream) = (curd & sour cream) / curd = 20 / 50 = 0.8
```
从结果来看，如果顾客购买了curd，那么购买 sour cream 的可能性为80%。

与支持度一样，置信度也在0~1之间，但是置信度是不对称的，也就是说 curd -> sour cream ≠ sour cream -> curd。

提升度，也被称为Lift，是一个衡量关联规则中X与Y之间相关性的重要指标。它表示在包含X的事务中同时包含Y事务的比例与单独看Y事务发生的比例之比。换句话说，提升度反映了商品A的出现对商品B的出现概率提升的程度。

提升度的具体计算公式为：Lift= ( P (A&B)/P (A))/P (B)=P (A&B)/P (A)/P (B)。在这个公式中，P(A&B)代表同时包含A和B的事务在所有事务中的比例，P(A)和P(B)分别代表单独包含A或B的事务在所有事务中的比例。

根据提升度的值，我们可以判断X和Y之间的关联关系。如果提升度大于1，那么说明X和Y存在正相关性，且越高表明正相关性越强；如果提升度小于1，那么说明X和Y存在负相关性，且越低表明负相关性越强；如果提升度等于1，那么说明X和Y不存在线性关联，即它们之间是相互独立的。例如，如果我们有一个关联规则：“购买了尿布”就“购买了啤酒”，通过计算提升度，我们可以量化这种关联关系的强度。的强度。

提升读的范围为0到无穷大。提升读越大，关联越强，提升读大于1表明先导与后继之间的关联比它们独立时候要来得更强，这意味着两件商品通常一起购买。

```
lift(sour cream -> curd) = support(curd & sour cream) / (support(curd) * support(sour cream))
                        = 0.2 / (0.25 * 0.3) = 2.66
```

从计算结果来看，当顾客购买凝乳的时候，同时购买酸奶油的期望值增加了166%。

# Apriori算法

Apriori算法是一种数据挖掘技术，主要用于发现频繁项集以及它们之间的关联规则。该算法的核心思想是通过迭代搜索找出数据库中的频繁项集，这一过程主要由连接（类矩阵运算）和剪枝（去除那些非必要的中间结果）两个步骤组成。

为了衡量数据项集的频繁程度，Apriori算法引入了支持度和置信度两个重要参数。支持度可以理解为一个项集在数据集中出现的频率，而置信度则反映了当一个数据项出现时，另一个数据项也出现的概率。

此外，Apriori算法还使用了提升度作为评判标准，以量化数据项集之间的关联关系强度。如果某一数据项集的出现会导致另一数据项集的出现概率显著增加，那么就认为这两项集之间存在强烈的关联关系。

可以使用第三方库： pip3 install mlxtend

首先创建一个数据集：

````python
transactions = [
    ['curd', 'sour cream'], ['curd', 'orange', 'sour cream'], 
    ['bread', 'cheese', 'butter'], ['bread', 'butter'], ['bread', 'milk'],
    ['apple', 'orange', 'pear'], ['bread', 'milk', 'eggs'], ['tea', 'lemon'],
    ['curd', 'sour cream', 'apple'], ['eggs', 'wheat flour', 'milk'],
    ['pasta', 'cheese'], ['bread', 'cheese'], ['pasta', 'olive oil', 'cheese'],
    ['curd', 'jam'], ['bread', 'cheese', 'butter'],
    ['bread', 'sour cream', 'butter'], ['strawberry', 'sour cream'],
    ['curd', 'sour cream'], ['bread', 'coffee'], ['online', 'garlic']
]
````

接着实现算法：

````python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

encoder = TransactionEncoder()
encoded_array = encoder.fit(transactions).transform(transactions)
df_itemsets = pd.DataFrame(encoded_array, columns = encoder.columns_)
````

这段代码使用了pandas库来处理数据，并利用了mlxtend库中的TransactionEncoder类来进行事务数据的编码。首先，创建了一个TransactionEncoder对象encoder。然后，通过调用fit方法对事务数据进行编码，并将编码后的结果存储在encoded_array中。最后，将编码后的数据转换为DataFrame格式，列名为编码器生成的列名，存储在df_itemsets中。


````python
df_itemsets
````

进一步的确认：

````python
print('Number of transactions: ', len(transactions))
````

识别频繁项集：

````python
from mlxtend.frequent_patterns import apriori
frenquent_itemsets = apriori(df_itemsets, min_support = 0.1, use_colnames = True)
````

查看结果：

````python
frenquent_itemsets
````

一个商品集可以由一件或者多件商品组成。实际上，apriori()已经返回了多个仅包含单件商品的商品集。

````python
frenquent_itemsets['length'] = frenquent_itemsets['itemsets'].apply(lambda itemset: len(itemset))
````

````python
print(frenquent_itemsets[frenquent_itemsets['length'] >= 2])
````

上面是不包含单件商品的商品集。

生成关联规则。

Apriori算法的第二步是为了这些商品集生成关联规则。

````python
from mlxtend.frequent_patterns import association_rules
rules = association_rules(frenquent_itemsets, metric = 'confidence', min_threshold = 0.5)
````

rules = association_rules(frenquent_itemsets, metric = 'confidence', min_threshold = 0.5): 这是调用association_rules函数并传递参数的语句。该函数用于生成关联规则。

- frenquent_itemsets: 这是输入参数，表示频繁项集的数据结构或数据集。频繁项集是指在事务数据中出现次数较多的项集。

- metric = 'confidence': 这是另一个输入参数，指定了关联规则的度量标准。在这里，度量标准被设置为置信度（confidence）。置信度是一种衡量关联规则可靠性的指标，它表示当一个项集中的项同时出现在另一个项集中时，这个关联规则成立的概率有多大。

- min_threshold = 0.5: 这是第三个输入参数，指定了最小阈值。关联规则的置信度必须大于等于这个阈值才会被保留。在这个例子中，最小阈值被设置为0.5。

最终，这段代码将生成满足置信度大于等于0.5的关联规则，并将结果存储在变量rules中。

````python
rules
````

可以看到，有些关联规则似乎是多余的。例如，同时存在 bread->butter 和 butter->bread。同行，基于(bread, cheese, butter)商品集也生成了一些关联规则，部分原因是因为置信度是不对称的。

可视化关联规则。

````python
rules_plot = pd.DataFrame()
rules_plot['antecedents'] = rules['antecedents'].apply(lambda x: ','.join(list(x)))
rules_plot['consequents'] = rules['consequents'].apply(lambda x: ','.join(list(x)))

rules_plot['lift'] = rules['lift'].apply(lambda x: round(x, 2))
````

接下来，将心创建的rules_plot数据框转换为一个矩阵，沿着水平方向排列后续，沿垂直方向排列先导。

````python
pivot = rules_plot.pivot(index = 'antecedents', columns = 'consequents', values = 'lift')
````

打印pivot数据：

````python
pivot
````

pivot数据框包含构建热图所需的所有元素。

````python
antecedents = list(pivot.index.values)
consequents = list(pivot.columns)

import numpy as np
pivot = pivot.to_numpy()
````

然后绘制热力图：

````python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots()

im = ax.imshow(pivot, cmap = 'Reds')
ax.set_xticks(np.arange(len(consequents)))
ax.set_yticks(np.arange(len(antecedents)))
ax.set_xticklabels(consequents)
ax.set_yticklabels(antecedents)

plt.setp(ax.get_xticklabels(), rotation = 4, ha = 'right', rotation_mode = 'anchor')

for i in range(len(antecedents)):
    for j in range(len(consequents)):
        if not np.isnan(pivot[i, j]):  # 添加了缺失的右括号
            text = ax.text(j, i, pivot[i, j], ha = 'center', va = 'center')

ax.set_title('Lift metric for frequent itemsets')
fig.tight_layout()
plt.show()
````
